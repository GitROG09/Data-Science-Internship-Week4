```python
import pandas as pd

# Create sample weather data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Temperature (°C)': [15, 18, 22, 27, 31, 34, 33, 32, 29, 25, 20, 16],
    'Rainfall (mm)': [30, 25, 40, 60, 100, 180, 220, 200, 150, 80, 50, 35],
    'Humidity (%)': [68, 65, 60, 55, 58, 70, 80, 82, 75, 68, 65, 67]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['Month'], df['Temperature (°C)'], marker='o', color='orange')
plt.title('Monthly Temperature Trend')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

#Insight:
The temperature gradually increases from January to June, peaking in summer, then cools down towards December.

```python
plt.bar(df['Month'], df['Rainfall (mm)'], color='skyblue')
plt.title('Monthly Rainfall Distribution')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.show()
```

#Insight:
Rainfall is highest during July and August, marking the monsoon season.

```python
avg_humidity = df['Humidity (%)'].mean()
plt.pie(df['Humidity (%)'], labels=df['Month'], autopct='%1.1f%%', startangle=90)
plt.title('Humidity Percentage by Month')
plt.show()
```

#Insight:
Humidity levels are highest during the monsoon months, reflecting the pattern of rainfall.

```python
fig, ax1 = plt.subplots()

# Temperature line (left axis)
ax1.plot(df['Month'], df['Temperature (°C)'], color='red', marker='o', label='Temperature (°C)')
ax1.set_xlabel('Month')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Second Y-axis for Rainfall
ax2 = ax1.twinx()
ax2.bar(df['Month'], df['Rainfall (mm)'], alpha=0.5, color='blue', label='Rainfall (mm)')
ax2.set_ylabel('Rainfall (mm)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Title and layout
plt.title('Temperature vs Rainfall (Monthly Comparison)')
fig.tight_layout()
plt.show()
```

#Insight:
The combined chart shows that as temperature rises until June, rainfall remains low. However, from July onwards, rainfall increases sharply while temperature begins to decline — clearly highlighting the seasonal monsoon shift.
